```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
a=pd.read_csv('automobile.csv')
```

```python
a.info()
```

```python
a.describe()
```

```python
a.head()
```

```python
sns.distplot(a['normalized_losses'])
plt.show()
```

```python
sns.distplot(a['city_mpg'],kde=False,rug=True)
plt.show()
```

```python
sns.jointplot(a['engine_size'],a['horsepower'])
plt.show()
```

```python
sns.jointplot(a['engine_size'],a['horsepower'],kind='hex')
plt.show()
```

```python
sns.jointplot(a['engine_size'],a['horsepower'],kind='kde')
plt.show()
```

```python
sns.pairplot(a[['normalized_losses','engine_size','horsepower']])
```

```python
sns.stripplot(a['fuel_type'],a['horsepower'])
plt.show()    
```

```python
sns.stripplot(a['fuel_type'],a['horsepower'],jitter=True)
```

```python
sns.swarmplot(a['fuel_type'],a['horsepower'])
```

```python
sns.boxplot(a['number_of_doors'],a['horsepower'])
```

```python
sns.boxplot(a['number_of_doors'],a['horsepower'],hue=a['fuel_type'])

# lines are the quartiles 25th, 50th 75th. The range is between max and min
the distance between 75th and the outlier is greater than 1.5 times the inter quartile range ie Q3-Q1
the data is possitively sqewed.
# hue argument can only have a categorical data
```

```python
            categorical data   numerical data 
sns.barplot(a['body_style'],a['horsepower'],hue=a['engine_location'])
# the lines above the bars are the confidence intervals for that given data in the bar
```

```python
sns.countplot(a['body_style'])
# gives a count of value chosen 
```

```python
sns.countplot(a['body_style'],hue=a['number_of_doors'])
```

```python
sns.pointplot(a['fuel_system'],a['horsepower'],hue=a['number_of_doors'])
```

```python
sns.lmplot(x='horsepower',y='peak_rpm',data=a)
# the light blue expanding line is the confidence interval 
```

```python
sns.lmplot(x='horsepower',y='peak_rpm',hue='fuel_type',data=a)
```

```python
sns.regplot(x='horsepower',y='peak_rpm',data=a)
```
